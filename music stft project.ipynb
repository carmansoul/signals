```python
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment

# Open the audio file with pydub
audio_file = AudioSegment.from_file('C:/Users/carma/Documents/REAPER Media/outter_space.mp3', format='mp3')

# Convert the audio file to raw data (bytes)
audio_data = audio_file.raw_data

# Convert the raw data to a 1-D array-like object
audio_data = np.frombuffer(audio_data, dtype=np.int16)

# Set the frame length and hop length for the STFT
frame_length = 1024
hop_length = 512

# Compute the short-time Fourier transform (STFT)
spectrogram, freqs, times, im = plt.specgram(audio_data, NFFT=frame_length, Fs=audio_file.frame_rate, noverlap=frame_length-hop_length)

# Display the waterfall plot of the STFT
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Waterfall plot of STFT')
plt.colorbar(im)
plt.show()
```
